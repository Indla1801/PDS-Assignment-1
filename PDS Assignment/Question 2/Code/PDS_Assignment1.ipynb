```python
# Create a histogram of math scores
plt.figure(figsize=(8, 8))
sns.histplot(df["math score"], bins=10, kde=True, color="purple")
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

The above graph tells the frequency distribution of math scores.The spread indicates how scores are distributed for different ranges. The higher concentration of students score is between 60 and 80. We can also see that very less students score below 40 or above 90. This visualization makes it easy to see how math scores are distributed.

```python
# Calculate mean scores by gender
avg_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()

# Plot the bar chart
avg_scores.plot(kind="bar", figsize=(8, 8), colormap="viridis")
plt.title("Average Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.legend(["Math", "Reading", "Writing"])
plt.show()
```

The above bar chart indicates the comparision of the average scores of male and female studets across math, reading and writing subjects. We can see that females score more than men in reading and writng but males score more in maths than females. This visualization makes it easy to compare male and females stdent strengths in different subjects.

```python
# Boxplot for math scores by parental education
plt.figure(figsize=(10, 10))
sns.boxplot(x="parental level of education", y="math score", data=df, palette="Set2")
plt.xticks(rotation=0)
plt.title("Math Scores by Parental Education Level")
plt.xlabel("Parental Education Level")
plt.ylabel("Math Score")
plt.show()
```

The above box plot shows how the math scores vary across different levels of parental education. From this we can tell whether students with more educated parents will have higher or lower math scores. We can tell that students whose parents having higher education level have higher math score. This visualization makes it easy to analyze how parental education influence academic success.

```python
# Scatter plot for math vs writing scores
plt.figure(figsize=(8, 8))
sns.scatterplot(x=df["math score"], y=df["reading score"], alpha=0.7)
plt.title("Math vs. Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

From the above scatter plot we can see the relationship between math and reading scores. If there is a positive correlation we can notice a trend where higher math scores align with higher writing scores. We can see that there is a positive correlation between math and reading scores. This visualization makes it easy to see relationship between subjects and predict the performances.

```python
# Compute correlation matrix
corr_matrix = df[["math score", "reading score", "writing score"]].corr()

# Plot the heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Subject Scores")
plt.show()
```

The above heatmap tell the relation between math, reading and writing scores.Strong positive correlations (closer to +1) indicates that the students perform well in one subject tend to perform well in others. The red color indicates strong positive correlations and the blue color indicates weak or negative correlations. We can see that reading and writing are closely linked but math is independent. This visualization makes it easier to analyze subject dependencies and predict performance among the subjects.
